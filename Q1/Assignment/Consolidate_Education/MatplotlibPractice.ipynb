# PIE CHARTS
##### https://matplotlib.org/api/_as_gen/matplotlib.pyplot.pie.html

```
labels= ["Facebook","Twitter","Youtube","Instagram"]
views = [340,200,430,329]
plt.pie(views,labels=labels)
```

#### If we apply less then 1

```
labels= ["Facebook","Twitter","Youtube","Instagram"]
views = [0.1,0.2,0.1,0.5]
plt.pie(views,labels=labels)


# You see pie chart automatically show 100% Graph
```

####Explode And Percentage in pie chart

```
labels= ["Facebook","Twitter","Youtube","Instagram"]
views = [340,200,430,329]
explode = [0,0,0,0.05]
plt.pie(views,labels=labels,explode=explode,autopct="%1.1f%%")
```

####Shadow And Wedgeprops in pie chart

```
labels= ["Facebook","Twitter","Youtube","Instagram"]
views = [340,200,430,329]
explode = [0,0,0,0.05]
plt.pie(views,labels=labels,explode=explode,autopct="%1.1f%%",shadow=True,wedgeprops={"width":0.30}) # Wedge property take a dicinory and it show 30 % pie out of 100
```

# Text Annotation
##### https://matplotlib.org/gallery/pyplots/annotation_basic.html#sphx-glr-gallery-pyplots-annotation-basic-py

```
views = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
x = range(1,14)

y_min = min(views)
x_index = x[views.index(y_min)]

plt.plot(x,views,label="Youtube Viewers")
plt.xlabel("Days") # Show x axis label
plt.ylabel("Views") # Show y axis label
plt.legend(loc="upper right") #Label Youtube Viewers is assign uper right
plt.grid(True,linestyle="-.")
plt.title("Youtube View on a daily basic") # Title show what the plot is about
plt.annotate('Lowest Values',
             xy=(x_index,y_min),
             xytext=(12,90),
             arrowprops=dict(facecolor="black",shrink=0.025))
plt.show()
```

# Plot using stylesheet
###### https://matplotlib.org/3.1.1/gallery/style_sheets/style_sheets_reference.html

```
views = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
x = range(1,14)

y_min = min(views)
x_index = x[views.index(y_min)]

plt.style.use("Solarize_Light2") # Above style is used to change sheet style
plt.plot(x,views,label="Youtube Viewers")
plt.xlabel("Days") # Show x axis label
plt.ylabel("Views") # Show y axis label
plt.legend(loc="upper right") #Label Youtube Viewers is assign uper right
plt.grid(True,linestyle="-.")
plt.title("Youtube View on a daily basic") # Title show what the plot is about
plt.annotate('Lowest Values',
             xy=(x_index,y_min),
             xytext=(12,90),
             arrowprops=dict(facecolor="black",shrink=0.025))
plt.show()
```
